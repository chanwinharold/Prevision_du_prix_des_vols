# Analyse Statistique

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("../data/data_cleaned.csv")
data.head()
```

### Étude Univariée

#### Variables Qualitatifs

```python
def add_labels(k, y):
    for i in range(len(k)):
        plt.text(i, y.iloc[i], y.iloc[i], horizontalalignment="center", verticalalignment="bottom")

for x in ["airline", "dep_time", "from", "time_taken", "dep_time", "arr_time", "to", "class"]:
    print("-"*50)
    print("Variable :", x)
    print("Mode :", data[x].mode()[0])
    table = pd.DataFrame()
    table["effectif"] = data[x].value_counts()
    table["frequency"] = data[x].value_counts(normalize=True)
    display(table)

    plt.pie(x=table["frequency"], labels=table.index, autopct="%.2f%%")
    plt.show()

    add_labels(table.index, table["effectif"])
    plt.bar(x=table.index, height=table["effectif"])
    plt.show()
```

#### Variable Quantitatif

```python
print("Moyenne: ", data["price"].mean().round(3))
print("Écart-type: ", data["price"].std().round(3))
print("Médiane: ", data["price"].median())

plt.hist(data["price"])
plt.show()
plt.boxplot(data["price"], showfliers=False, showmeans=True)
plt.show()
```
